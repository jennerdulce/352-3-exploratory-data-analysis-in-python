```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weightsDataFrame = pd.read_csv('./weights.tsv', header=None)
weights_lbs = pd.Series(weightsDataFrame[0].values)
```

```python
weights_kg = (weights_lbs * 0.453592).round(2)
```

```python
print(f"Weights in Pounds:")
weights_pounds_mean = weights_lbs.mean()
weights_pounds_median = weights_lbs.median()
weights_pounds_stdn = weights_lbs.std()
print(f"Mean: {weights_pounds_mean:.2f} lbs, Median: {weights_pounds_median:.2f} lbs, Standard Deviation: {weights_pounds_stdn:.2f} lbs")

print(f"Weights in Kilograms:")
weights_kg_mean = weights_kg.mean()
weights_kg_median = weights_kg.median()
weights_kg_stdn = weights_kg.std()
print(f"Mean: {weights_kg_mean:.2f} kg, Median: {weights_kg_median:.2f} kg, Standard Deviation: {weights_kg_stdn:.2f} kg\n")
```

```python
plt.figure(figsize=(10, 6))
plt.hist(weights_kg, bins=10, edgecolor='black')
plt.xlabel('Weight (in kilograms)')
plt.ylabel('Frequency')
plt.title('Histogram of Weights (in kilograms)')
plt.savefig('./1_4_weight_in_kg_histogram.png', dpi=300, bbox_inches='tight')
plt.close()
```
